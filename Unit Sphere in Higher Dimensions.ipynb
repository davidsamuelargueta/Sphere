The unit sphere is simply a sphere with a radius of 1. As we know the volume of sphere given by
<center>$V = \frac{4}{3}\pi R^3$, where R is the radius of the sphere</center>
thus the volume of a unit sphere is $\frac{4}{3}\pi â‰ˆ 4.19$


The Volume of a sphere in n dimensions with radius R, where R $\in\mathbb{R}^{+}$ can be given by <br>
<center>$V_n(R) = \begin{cases} 
      1 & \text{if n = 0} \\
      2R & \text{if n = 1} \\
      \frac{2\pi R^2}{n}V_{n-2}(R) & \text{otherwise}
   \end{cases}$</center>

```python
#First import the required packages
import numpy as np
import math as math
from math import pi as pi
import matplotlib.pyplot as plt

#Create a function that represents the relation
def volume(n,R):
    R=1 #Let our radius be 1, for this case as the unit sphere has radius 1
    if n==0:
        return 1
    if n==1:
        return 2*R
    else:
        return (2*pi/n)*volume(n-2,R)*R**2
```

```python
#By setting n=2, we are given the area of the unit circle
volume(2,1)
```

```python
#Setting n=3, we have the volume of the unit sphere
volume(3,1)
```

We can also plot the volume function to see the relation between n and the volume of the unit sphere as we go to higher dimensions

```python
n = np.arange(0, 30, 1)
```

```python
#Create an empty list called volumes so we can then append the different volumes for plotting
volumes = []
for i in range(len(n)):
   volumes.append(volume(n[i], 1))
```

```python
#Confirm we have our volumes, notice that as n increases the volumes approach 0
print(volumes)
```

```python
plt.plot(n, volumes)
plt.title("Volume of Sphere with Radius 1, in Higher Dimensions")
plt.xlabel("n")
plt.ylabel("$V_n(R)$")
```

So from our plot, we can see the relation between the volume and the number of dimensions of an n-Ball. Notice that when n=5, we reach the point where the volume is the greatest, as we continue into higher dimensions, the volumes gradually decrease and approach 0 as n approaches infinity, indicating there is an inverse relation between the two.
